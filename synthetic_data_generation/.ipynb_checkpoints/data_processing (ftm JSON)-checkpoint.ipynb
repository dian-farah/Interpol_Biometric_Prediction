```python
# functions to scrape origin API via Selenium
def make_driver(headless: bool = False):
    options = Options()
    options.add_argument("--disable-gpu")
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")

    # If you want headless:
    if headless:
        # Chrome 109+ recommended headless mode
        options.add_argument("--headless=new")

    return webdriver.Chrome(
        service=Service(ChromeDriverManager().install()),
        options=options,
    )

def get_notice_data(notice_url, driver):
    """
    Directly accesses the Interpol Notice API link and returns the JSON.
    """
    try:
        driver.get(notice_url)
        # Find the JSON text rendered in the browser window
        soup = BeautifulSoup(driver.page_source, "html.parser")
        pre = soup.find("pre")
        
        if pre and pre.text:
            return json.loads(pre.text)
        return None
    except Exception as e:
        print(f"Error fetching {notice_url}: {e}")
        return None

def worker(chunk_df, worker_id):
    driver = make_driver(headless=False)
    results = []
    
    for _, row in tqdm(chunk_df.iterrows(), total=len(chunk_df), desc=f"Worker {worker_id}", position=worker_id):
        person_id, url, num_apis = row[0], row[1], row[2]
        try:
            driver.get(url)
            soup = BeautifulSoup(driver.page_source, "html.parser")
            pre = soup.find("pre")
            if pre and pre.text:
                data = json.loads(pre.text)
                df_temp = pd.json_normalize(data)
                df_temp['id'] = person_id
                df_temp['num apis'] = num_apis
                results.append(df_temp)
        except Exception as e:
            continue 
            
    driver.quit()
    return pd.concat(results) if results else pd.DataFrame()

# url = 'https://ws-public.interpol.int/notices/v1/red/2019-108810'
# get_notice_data(url, make_driver())
```

```python
import ast

def process_list_string(s):
    if pd.isna(s): # Handles actual NaN values in the column
        return None
    try:
        # Try evaluating after replacing 'nan' with 'None'
        return ast.literal_eval(s.replace('nan', 'None'))
    except (ValueError, SyntaxError):
        # Handle cases where the string is not a list format, e.g., a single value
        return s

origin_df['nationalities'] = origin_df['nationalities'].apply(process_list_string)
origin_df['eyes_colors_id'] = origin_df['eyes_colors_id'].apply(process_list_string)
origin_df['arrest_warrants'] = origin_df['arrest_warrants'].apply(process_list_string)
origin_df['hairs_id'] = origin_df['hairs_id'].apply(process_list_string)
# origin_df['languages_spoken_ids'] = origin_df['languages_spoken_ids'].apply(process_list_string)
origin_df['_embedded.links'] = origin_df['_embedded.links'].apply(process_list_string)
```

```python
cols_to_explode = ['nationalities', 'eyes_colors_id', 'arrest_warrants', 'hairs_id', '_embedded.links']

for col in cols_to_explode:
    origin_df = origin_df.explode(col)
```

```python
origin_df = origin_df.reset_index(drop=True)
```

```python
def normalise_data(df, col):
    normalised = pd.json_normalize(df)
    normalised = normalised.add_prefix('(' + col +')')
    return normalised

origin_normalise = pd.concat([origin_df, normalise_data(origin_df['arrest_warrants'], 'arrest_warrants')], axis = 1)
origin_normalise = pd.concat([origin_normalise, normalise_data(origin_normalise['_embedded.links'], 'embedded_links')], axis= 1)
```

```python
import pandas as pd
from dateutil.relativedelta import relativedelta
from datetime import date

def calculate_detailed_age(born):
    # Check if 'born' is NaT or None
    if pd.isnull(born):
        return None
    
    today = date.today()
    # Ensure 'born' is a date object, as pandas timestamps can sometimes 
    # behave differently with relativedelta
    age = relativedelta(today, born.date())
    return age

# Your existing setup
origin_normalise['date_of_birth'] = pd.to_datetime(
    origin_normalise['date_of_birth'], 
    format='%Y/%m/%d', 
    errors='coerce'
)

# Apply with the updated function
origin_normalise['age'] = origin_normalise['date_of_birth'].apply(calculate_detailed_age)
```

```python
origin_normalise['age'] = origin_normalise['age'].apply(lambda x: x.years if x else None)
```
