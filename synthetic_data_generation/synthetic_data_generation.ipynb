```python
def process_list_string(s):
    if pd.isna(s): # Handles actual NaN values in the column
        return None
    try:
        # Try evaluating after replacing 'nan' with 'None'
        return ast.literal_eval(s.replace('nan', 'None'))
    except (ValueError, SyntaxError):
        # Handle cases where the string is not a list format, e.g., a single value
        return s

def normalise_data(df, col):
    normalised = pd.json_normalize(df)
    normalised = normalised.add_prefix('(' + col +')')
    return normalised

def calculate_detailed_age(born):
    # Check if 'born' is NaT or None
    if pd.isnull(born):
        return None
    
    today = date.today()
    # Ensure 'born' is a date object, as pandas timestamps can sometimes 
    # behave differently with relativedelta
    age = relativedelta(today, born.date())
    return age
```

## Generating Synthetic Client Data (To Train & Test Models)
Total Records: 10,000
- Matches (Sanctioned): 500 (5% of the data)
- Non-matches (Clean): 9,500 (95% of the data) - reflecting real-world scenarios where most clients are not sanctioned.
- Target Column: match_label (1 = Sanctioned, 0 = Clean).

The synthetic data includes only typical client-facing features, excluding internal sanction descriptions and crime details:
- client_id: A unique identifier (e.g., C000001).
- full_name: Realistic names derived and shuffled from the original dataset.
- date_of_birth: Formatted as YYYY-MM-DD.
- gender: M/F/Other.
- nationality_code: Two-letter country codes (e.g., ru, tj, co).
- height_cm & weight_kg: Numerical biometric data.
- hair_color & eye_color: Categorical biometric descriptors.
- spoken_languages: Three-letter language codes.

Key Considerations:
- Synthetic data maintains biometric sparsity (e.g. 50% of original data lacked hair colour and eye colour, which is reflected
- For the 500 matches, minor variations were introduced (e.g. name misspelling, height difference, date of birth alteration) to reflect real world data

```python
# Load the dataset
df = pd.read_csv('processed/mega_sanctions_final.csv')

# Inspect the columns and the first few rows
print(df.columns.tolist())
print(df.head())
print(df.info())

# Identify columns suitable for client data
# name, birth_date, hairColor, height, eyeColor, GENDER, countries (maybe map to nationality), weight, languages_spoken_ids
client_cols = ['name', 'birth_date', 'hairColor', 'height', 'eyeColor', 'GENDER', 'countries', 'weight', 'languages_spoken_ids']

# Check missingness
missing_stats = df[client_cols].isnull().mean()
print("\nMissingness in client columns:")
print(missing_stats)
```

```python
# Select client columns
client_cols = ['name', 'birth_date', 'hairColor', 'height', 'eyeColor', 'GENDER', 'countries', 'weight', 'languages_spoken_ids']

# Extract distributions for non-matches
distributions = {}
for col in client_cols:
    if col != 'name':
        # Get counts including NaNs
        distributions[col] = df[col].value_counts(dropna=False, normalize=True)

# Function to generate synthetic names for non-matches
# We can split existing names and shuffle them.
names = df['name'].dropna().unique()
first_names = [n.split()[0] for n in names if len(n.split()) > 0]
last_names = [n.split()[-1] for n in names if len(n.split()) > 1]

def generate_random_name():
    fn = np.random.choice(first_names)
    ln = np.random.choice(last_names)
    return f"{fn} {ln}"

# Check distributions
print(distributions['GENDER'])
print(distributions['hairColor'].head())
```

```python
# Number of rows
n_total = 10000
n_matches = 500 # 5% of data will be matched to sanctions
n_non_matches = n_total - n_matches

# 1. GENERATE MATCHES
# Select 500 random records from the sanctions list
matches_orig = df.sample(n=n_matches, random_state=42)

# Select client-relevant columns
client_map = {
    'name': 'full_name',
    'birth_date': 'date_of_birth',
    'GENDER': 'gender',
    'countries': 'nationality_code',
    'height': 'height_cm',
    'weight': 'weight_kg',
    'hairColor': 'hair_color',
    'eyeColor': 'eye_color',
    'languages_spoken_ids': 'spoken_languages'
}

matches = matches_orig[list(client_map.keys()) + ['id']].rename(columns=client_map)
matches = matches.rename(columns={'id': 'matched_sanction_id'})

# For half of the matches, let's introduce some variability to simulate real-world data
def introduce_fuzzy(row):
    """
    Role the dice to choose whether we create variability for name, height, or birth date
    The names will be altered through random swapping of characters for names length > 5
    The date of birth will be altered at either year, month, day or mixed:
        - year: 1 to 3
        - month: 1 to 6
        - day: 1 to 20
    The height will be altered by +- 1 to 5cm if height is not null
    """
    # Misspell name slightly
    name = str(row['full_name'])
    if len(name) > 5:
        strategy = random.choice(['swap', 'shuffle_slice', 'drop', 'duplicate'])
        char_list = list(name)
        
        if strategy == 'swap':
            idx = random.randint(0, len(char_list) - 2)
            char_list[idx], char_list[idx+1] = char_list[idx+1], char_list[idx]
            
        elif strategy == 'shuffle_slice':
            slice_len = min(random.randint(3, 4), len(char_list))
            start_idx = random.randint(0, len(char_list) - slice_len)
            slice_chars = char_list[start_idx:start_idx + slice_len]
            random.shuffle(slice_chars)
            char_list[start_idx:start_idx + slice_len] = slice_chars
            
        elif strategy == 'drop':
            char_list.pop(random.randint(0, len(char_list) - 1))
            
        elif strategy == 'duplicate':
            idx = random.randint(0, len(char_list) - 1)
            char_list.insert(idx, char_list[idx])
        
        row['full_name'] = "".join(char_list)
        
     # Modify height by +/- 1 to 5
    if pd.notnull(row['height_cm']) and row['height_cm'] > 0:
        row['height_cm'] += random.randint(-5, 5)
    
    # Modify age anywhere between 1 to 3 years, 1 to 6 months, 1 to 20 days
    if pd.notnull(row['date_of_birth']):
        try:
            dt = pd.to_datetime(row['date_of_birth'])
            
            # Pick a random shift type
            shift_type = random.choice(['year', 'month', 'day', 'mix'])
            
            if shift_type == 'year':
                # Shift 1 to 3 years
                dt += relativedelta(years=random.randint(1, 3) * random.choice([-1, 1])) # change from 10 years to 3 years (+/-)
            elif shift_type == 'month':
                # Shift 1 to 6 months
                dt += relativedelta(months=random.randint(1, 6) * random.choice([-1, 1]))
            elif shift_type == 'day':
                # Shift 1 to 20 days
                dt += relativedelta(days=random.randint(1, 20) * random.choice([-1, 1]))
            else:
                # A mix of small changes to simulate general data entry error
                dt += relativedelta(
                    years=random.randint(0, 2) * random.choice([-1, 1]),
                    months=random.randint(0, 3) * random.choice([-1, 1]),
                    days=random.randint(1, 10) * random.choice([-1, 1])
                )
            
            row['date_of_birth'] = dt.strftime('%Y-%m-%d')
        except Exception:
            pass # Fallback in case of conversion errors
    return row

matches = matches.apply(introduce_fuzzy, axis=1)
matches['match_label'] = 1

# 2. GENERATE NON-MATCHES
# Define non-match data
non_matches_data = {}

# Sample random IDs from original sanctions list for non-matches
matches_id = set(matches_orig)
df_non_matches = df[~df['id'].isin(matches_id)]
non_matches_data['matched_sanction_id'] = df_non_matches['id'].sample(n=n_non_matches, replace=True, random_state=7).values

# display(non_matches_data)

# Distribution sampling function
def sample_from_dist(col_name, size):
    """
    Helper function to randomly sample data from original sanctions dataset for the non-matched clients.
    The rest of the unsampled data will have random generated names
    """
    dist = distributions[col_name]
    values = dist.index.tolist()
    probs = dist.values.tolist()
    # Handle NaNs properly in choice
    # values can contain np.nan which is fine for np.random.choice if we handle the list correctly
    sampled = np.random.choice(values, size=size, p=probs)
    return sampled

for orig_col, new_col in client_map.items():
    if orig_col == 'name':
        # Generate random names
        non_matches_data[new_col] = [generate_random_name() for _ in range(n_non_matches)]
    else:
        non_matches_data[new_col] = sample_from_dist(orig_col, n_non_matches)

non_matches = pd.DataFrame(non_matches_data)
non_matches['match_label'] = 0

# Ensure non-matches aren't actually in the sanctions list by checking names
sanctioned_names = set(df['name'].str.upper().tolist())
# If a generated name is in sanctioned_names, regenerate it
non_matches['full_name'] = non_matches['full_name'].apply(
    lambda x: generate_random_name() if str(x).upper() in sanctioned_names else x
)

# 3. COMBINE AND SHUFFLE
synthetic_df = pd.concat([matches, non_matches], ignore_index=True)
synthetic_df = synthetic_df.sample(frac=1, random_state=42).reset_index(drop=True)

# Add a synthetic client_id
synthetic_df['client_id'] = ['C' + str(i).zfill(6) for i in range(1, n_total + 1)]

# Reorder columns
final_cols = ['client_id', 'full_name', 'date_of_birth', 'gender', 'nationality_code',
              'height_cm', 'weight_kg', 'hair_color', 'eye_color', 'spoken_languages', 'match_label', 'matched_sanction_id']
synthetic_df = synthetic_df[final_cols]

display(synthetic_df)

# Save to CSV
synthetic_df.to_csv('synthetic_client_data.csv', index=False)

# Quick check on distributions
print("\nSynthetic data missingness:")
print(synthetic_df.isnull().mean())
print("\nMatch count:")
print(synthetic_df['match_label'].value_counts())
print("\nFirst 10 rows:")
print(synthetic_df.head(10))
```

```python
# Debugging missingness
print("Data types in synthetic_df:")
print(synthetic_df.dtypes)

# Check if 'nan' is a string
print("\nIs 'nan' a string in hair_color?")
print(synthetic_df['hair_color'].apply(lambda x: str(x).lower() == 'nan').mean())

# Fix the distribution sampling to ensure NaN is handled correctly as a real null
def sample_from_dist_fixed(col_name, size):
    dist = distributions[col_name]
    values = dist.index.tolist()
    probs = dist.values.tolist()
    # Replace any string 'nan' or actual nan with a placeholder if needed,
    # but the best way is to keep them as objects and then ensure they are np.nan
    sampled = np.random.choice(len(values), size=size, p=probs)
    return [values[i] for i in sampled]

# Re-generate non-matches with the fix
for orig_col, new_col in client_map.items():
    if orig_col == 'name':
        pass # already done
    else:
        non_matches[new_col] = sample_from_dist_fixed(orig_col, n_non_matches)

# Re-combine
synthetic_df = pd.concat([matches, non_matches], ignore_index=True)
synthetic_df = synthetic_df.sample(frac=1, random_state=42).reset_index(drop=True)
synthetic_df['client_id'] = ['C' + str(i).zfill(6) for i in range(1, n_total + 1)]
synthetic_df = synthetic_df[final_cols]

# Final verification
print("\nCorrected Synthetic data missingness (isnull()):")
print(synthetic_df.isnull().mean())

# Save again
synthetic_df.to_csv('synthetic_client_data.csv', index=False)
```
